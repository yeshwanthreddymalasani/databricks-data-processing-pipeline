```python
df = spark.read.csv("/Workspace/Users/yeshwanthreddy0974@gmail.com/databricks-data-processing-pipeline/Datasets/ecommerce_data.csv", header=True, inferSchema=True)

df.show()




```

```python
df.createOrReplaceTempView("ecommerce_data")
```

```python
%sql
SELECT category, SUM(price * quantity) AS total_revenue
FROM ecommerce_data
GROUP BY category;

SELECT product, SUM(quantity) AS total_sold
FROM ecommerce_data
GROUP BY product
ORDER BY total_sold DESC;
```
